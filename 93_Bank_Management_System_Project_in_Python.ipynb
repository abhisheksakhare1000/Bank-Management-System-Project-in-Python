```python
class sbi_bank:
    #generic data
    address = 'abc'
    phone_no = '9373834313'
    bank_maneger = 'Xyz'
    IFSC_code = '1224677'
    
    #specific data
    def  __init__(self):
        self.name = self.get_name()
        self.mobile = self.mobile()
        self.balance = self.balance()
        self.account = self.account()
    def display(self):
        #display specific data
        print("\nName of person:", self.name)
        print("the mobile:", self.mobile)
        print(" the balacne:", self.balance)
        print("Account number:", self.account)
        # display generic data
        print("bank info as follows:\n")
        print("address of bank:", self.address)
        print("Bank phone no:\n", self.phone_no)
        print("Bank Manager name:", self.bank_maneger)
        print("Bank IFSC code",self.IFSC_code)
    @staticmethod
    def get_name():
        return  input("Enter the name:")
    @staticmethod
    def mobile():
        return input("Enter the mobile:")
    @staticmethod
    def balance():
        return int(input("Enter the balance"))
    @staticmethod
    def account():
        return input("Enter the account No:")
    
    
    # change Generic data using class method
    
    @classmethod
    def change_address(cls, new_data):
        cls.address = new_data
    @classmethod
    def change_phone(cls,new_data):
        cls.phone_no = new_data
    @classmethod
    def change_bank_maneger(cls,new_data):
        cls.bank_maneger = new_data
    @classmethod
    def change_IFSC_code (cls, new_data):
        cls.IFSC_code = new_data
    
    # Chnage specific data 
    
    def change_name(self,new_data):
        self.name = new_data
    def change_mobile(self,new_data):
        self.mobile = new_data
    
    # Add bank bank balance
    def add_balance(self,new_data):
        self.balance =self.balance +new_data
    
    # Withdraw bank balance
    def withdraw_balance(self,new_data):
        self.balance =self.balance-new_data
        
```
